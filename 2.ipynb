```python
import secrets
import matplotlib.pyplot as plt

# Define o número de rodadas
num_rodadas = 5000

# Inicializa a pontuação do jogador
pontos = 0

# Lista para armazenar a evolução dos pontos
evolucao_pontos = []

# Loop para as rodadas do jogo
for rodada in range(num_rodadas):
    # Gera um número aleatório entre 0 e 1
    resultado = secrets.SystemRandom().random()

    # Se o resultado for menor que 0.5, considera cara (soma um ponto)
    if resultado < 0.5:
        pontos += 1
    # Caso contrário, considera coroa (subtrai um ponto)
    else:
        pontos -= 1

    # Adiciona a pontuação atual à lista de evolução dos pontos
    evolucao_pontos.append(pontos)

# Plota o gráfico da evolução dos pontos
plt.plot(evolucao_pontos)
plt.title('Evolução dos Pontos no Jogo de Cara ou Coroa')
plt.xlabel('Rodada')
plt.ylabel('Pontos')
plt.show()
```

```python
import secrets
import matplotlib.pyplot as plt
import numpy as np

# Define o número de rodadas
num_rodadas = 50000

# Inicializa a pontuação do jogador
pontos = 0

# Lista para armazenar a evolução dos pontos
evolucao_pontos = []

# Lista para armazenar o topo histórico acumulado (maior valor até cada rodada)
topo_historico = []

# Loop para as rodadas do jogo
for rodada in range(num_rodadas):
    # Gera um número aleatório entre 0 e 1
    resultado = secrets.SystemRandom().random()

    # Se o resultado for menor que 0.5, soma um ponto (cara)
    if resultado < 0.5:
        pontos += 1
    # Caso contrário, subtrai um ponto (coroa)
    else:
        pontos -= 1

    # Adiciona a pontuação atual à lista de evolução
    evolucao_pontos.append(pontos)

    # Atualiza o topo histórico (maior valor até agora)
    if rodada == 0:
        topo_historico.append(pontos)
    else:
        topo_historico.append(max(topo_historico[-1], pontos))

# Cria um array de rodadas para o eixo X
rodadas = np.arange(num_rodadas)

# Calcula a diferença entre o topo histórico e a pontuação atual
diferenca = np.array(topo_historico) - np.array(evolucao_pontos)

# Define as áreas onde a diferença é menor que 10
regiao_proxima = diferenca > 10

# Plota o gráfico da evolução dos pontos e do topo histórico
plt.plot(evolucao_pontos, label="Pontuação atual", color='blue')
plt.plot(topo_historico, 'r--', label="Topo histórico", linewidth=1.5)

# Preenche de verde transparente onde a diferença é < 10
plt.fill_between(
    rodadas,
    evolucao_pontos,
    topo_historico,
    where=regiao_proxima,
    color='red',
    alpha=0.3,
    label="Distância > 10 pontos"
)

plt.title('Evolução dos Pontos no Jogo de Cara ou Coroa')
plt.xlabel('Rodada')
plt.ylabel('Pontos')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
