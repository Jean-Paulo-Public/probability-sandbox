```python
import random
import matplotlib.pyplot as plt

# Define o nmero de rodadas
num_rodadas = 1000

# Inicializa a pontuao do jogador
pontos = 0

# Lista para armazenar a evoluo dos pontos
evolucao_pontos = []

# Loop para as rodadas do jogo
for rodada in range(num_rodadas):
    # Gera um nmero aleatrio entre 0 e 1
    resultado = random.random()

    # Se o resultado for menor que 0.5, considera cara (soma um ponto)
    if resultado < 0.5:
        pontos += 1
    # Caso contrrio, considera coroa (subtrai um ponto)
    else:
        pontos -= 1

    # Adiciona a pontuao atual  lista de evoluo dos pontos
    evolucao_pontos.append(pontos)

# Plota o grfico da evoluo dos pontos
plt.plot(evolucao_pontos)
plt.title('Evoluo dos Pontos no Jogo de Cara ou Coroa')
plt.xlabel('Rodada')
plt.ylabel('Pontos')
plt.show()
```

```python
import os
import matplotlib.pyplot as plt

# Define o nmero de rodadas
num_rodadas = 5000

# Inicializa a pontuao do jogador
pontos = 0

# Lista para armazenar a evoluo dos pontos
evolucao_pontos = []

# Loop para as rodadas do jogo
for rodada in range(num_rodadas):
    # Gera um nmero aleatrio entre 0 e 1 usando os.urandom
    random_bytes = os.urandom(8)  # Gera 8 bytes aleatrios
    resultado = int.from_bytes(random_bytes, byteorder='big') / (2**64 - 1)  # Converte para um nmero entre 0 e 1


    # Se o resultado for menor que 0.5, considera cara (soma um ponto)
    if resultado < 0.5:
        pontos += 1
    # Caso contrrio, considera coroa (subtrai um ponto)
    else:
        pontos -= 1

    # Adiciona a pontuao atual  lista de evoluo dos pontos
    evolucao_pontos.append(pontos)

# Plota o grfico da evoluo dos pontos
plt.plot(evolucao_pontos)
plt.title('Evoluo dos Pontos no Jogo de Cara ou Coroa')
plt.xlabel('Rodada')
plt.ylabel('Pontos')
plt.show()
```
